```python
import pandas as pd
import numpy as np
import os
```

```python
DATA_DIR="./../data/"
file_a=[]
for files in os.listdir(DATA_DIR):
    add=DATA_DIR+"/"+files
    for f_a in os.listdir(add):
        file_a.append(add+"/"+f_a)
```

```python
print (len(file_a))
```

```python
dataframe=pd.read_csv(file_a[0],delimiter="|")
```
